```python
import json

with open(r"C:\Users\MainUser\project\cs224w_cb_graph\text_data\logan.json", "r", encoding="utf-8") as f:
    speeches = json.load(f)

print(len(speeches))
speech_text = speeches[0]["text"]
print(speech_text)
```

```python
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
topic_model = BERTopic(embedding_model=model)

docs = [speech_text]  # your input speech or batch

if len(docs) > 1:
    topics, probs = topic_model.fit_transform(docs)
    topic_keywords = topic_model.get_topic_info()
else:
    # Fallback: no topic modeling possible
    topics, probs = [-1], [None]
    topic_keywords = None
    print("⚠️ Only one document provided — skipping BERTopic.")

```

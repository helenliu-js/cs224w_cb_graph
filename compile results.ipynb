```python
import pandas as pd
path = r"C:\Users\MainUser\Desktop\cs230_result.xlsx"
normal = pd.read_excel(path, sheet_name="Normal")
st = pd.read_excel(path, sheet_name="ST")
blogistic = pd.read_excel(path, sheet_name="Blogistic")
sst = pd.read_excel(path, sheet_name="SST")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.plot(normal["Epochs"], normal["Train Loss"], label="Normal - Train Loss", linewidth=2)
plt.plot(st["Epochs"], st["Train Loss"], label="Student-t - Train Loss", linewidth=2)
plt.plot(sst["Epochs"], sst["Train Loss"], label="Skewed Student-t - Train Loss", linewidth=2)
plt.plot(blogistic["Epochs"], blogistic["Train Loss"], label="B Logistic - Train Loss", linewidth=2)

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss Comparison")

# Legend outside the plot (bottom center)
plt.legend(
    loc="upper center",
    bbox_to_anchor=(0.5, -0.15),
    ncol=2,                # number of columns in legend (adjust as needed)
    frameon=False          # no box around legend
)

plt.savefig("train_loss_comparison.png", dpi=300, bbox_inches="tight")

plt.show()


```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# --- Plot Dev Loss and Confidence Intervals ---

plt.plot(normal["Epochs"], normal["Dev Loss"], label="Normal - Dev Loss", linewidth=2)

# Student-t
plt.plot(st["Epochs"], st["Dev Loss"], label="Student-t - Dev Loss", linewidth=2)

# Skewed Student-t
plt.plot(sst["Epochs"], sst["Dev Loss"], label="Skewed Student-t - Dev Loss", linewidth=2)

# B Logistic
plt.plot(blogistic["Epochs"], blogistic["Dev Loss"], label="B Logistic - Dev Loss", linewidth=2)


# --- Labels and Title ---
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Validation (Dev) Loss Comparison")

# --- Legend outside (bottom center) ---
plt.legend(
    loc="upper center",
    bbox_to_anchor=(0.5, -0.15),
    ncol=2,
    frameon=False
)

# --- Save the figure ---
plt.savefig("dev_loss_comparison.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
import requests

url = "https://fedinprint.org/api/author/l:logan-lorie/items"
params = {
    "limit": 10000
}
headers = {
    "x-api-key": API_KEY
}
response = requests.get(url, params=params, headers=headers)
data = response.json()
print(len(data["records"]))
```

```python
import json

def get_ids():
    with open("text_data/logan.json", "r", encoding="utf-8") as f:
        text_data = json.load(f)
        return list(map(lambda x: x["id"], text_data))

```

```python
ids = get_ids()
records = data["records"]
idx = 0

for record in records:

    if record["id"].split(":")[0].endswith("sp"):
        record_id = record["id"]
        if record_id in ids:
            continue

        print(record["id"])
        print(record["file"])
        idx += 1
print(idx)
```

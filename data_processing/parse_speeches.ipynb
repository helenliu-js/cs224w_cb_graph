```python
def force_single_line(t: str):
    if not isinstance(t, str):
        return t

    # Remove real newline characters
    t = t.replace("\n", " ").replace("\r", " ")

    # Remove escaped sequences like \n
    t = t.replace("\\n", " ")

    # Remove NBSP
    t = t.replace("NBSP", " ").replace("\xa0", " ")

    # Collapse multiple spaces
    t = " ".join(t.split())

    return t

author_short_name = data_processing_utils.get_author_short_name("l:logan-lorie")
filename = f"text_data/{author_short_name}.json"

# Load JSON
with open(filename, "r", encoding="utf-8") as f:
    data = json.load(f)

# Modify rows
for row in data:
    # Fix date if needed
    # row["date"] = fix_date(row["date"])

    # Flatten text into ONE long line
    row["text"] = force_single_line(row["text"])

# Save back
with open(filename, "w", encoding="utf-8") as f:
    json.dump(data, f, indent=2, ensure_ascii=False)

print("Cleaned & flattened JSON saved.")
```
